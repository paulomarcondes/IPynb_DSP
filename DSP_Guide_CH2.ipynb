### Mean and Standard Deviation
The mean is a simple summation of all elements, divided by the number of elements:

$$\mu = \frac{1}{N} \sum^{N-1}_{i=0} ~ x_i$$

*Equation 2-1 Calculation of a signal's mean. The signal is contained in* $x_0$ *through* $x_{N-1}$, $i$ *is an index that runs through these values and* $\mu$ *is the mean.*

$$\sigma^2 = \frac{1}{N-1} \sum^{N-1}_{i=0} ~ (x_i - \mu)$$

*Equation 2-2 Calculation of the standard deviation of a signal. The signal is stored in* $x_i$, $\mu$ *it the mean found from Eq. 2-1,* $N$ *is the number of samples and* $\sigma$ *is the standard deviation.*
